```python
# Your code here
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data_train = pd.read_csv(f'https://raw.githubusercontent.com/4GeeksAcademy/machine-learning-content/master/assets/clean_titanic_train.csv')
X_train= data_train.drop("Survived", axis = 1) 
y_train = data_train['Survived']
print(X_train, y_train)
data_test = pd.read_csv(f'https://raw.githubusercontent.com/4GeeksAcademy/machine-learning-content/master/assets/clean_titanic_test.csv')
X_test = data_test.drop('Survived', axis=1) 
y_test = data_test['Survived']
print(X_test, y_test)
```
